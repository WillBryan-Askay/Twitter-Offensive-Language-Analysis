However, measures should be taken to prevent overfitting, such as implementing early stopping, dropout, or regularization techniques.

### Architectures:

#### *Long Short-Term Memory*: 
The Long Short-Term Memory (LSTM) model is a type of recurrent neural network (RNN) designed to learn long-term dependencies in sequence data. Unlike standard feedforward neural networks, LSTM has feedback connections. It can not only process single data points (such as images), but also entire sequences of data (such as speech or video). The central concept behind LSTM is the cell state, which is a horizontal line running through the top of the LSTM unit, allowing information to flow across the sequence without being necessarily altered by the LSTM unit. This information flow is controlled by three types of gates (with sigmoid activation functin): the input gate, the forget gate, and the output gate.

In mathematical terms, LSTM can be described by the following equations:

Forget Gate: $f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)$
Input Gate: $i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)$
Candidate layer: $\widetilde{C}t = \tanh(W_C \cdot [h{t-1}, x_t] + b_C)$
Update of cell state: $C_t = f_t * C_{t-1} + i_t * \widetilde{C}_t$
Output Gate: $o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)$
Hidden state: $h_t = o_t * \tanh(C_t)$
Here, $\sigma$ denotes the sigmoid function, $*$ denotes element-wise multiplication, $W$ and $b$ are the weights and biases for each gate, and $[h_{t-1}, x_t]$ is the concatenation of the input vector and previous hidden state. The forget gate $f_t$ decides which information to discard from the cell state, the input gate $i_t$ decides which information is to be stored in the cell state, and the output gate $o_t$ decides what the next hidden state should be.

#### *Forward Feeding Neural Network*:
The Feed-Forward Neural Network (FFNN) is a type of artificial neural network wherein connections between the nodes do not form a cycle. In other words, the information moves in only one direction, forward, from the input nodes, through the hidden nodes (if any) and to the output nodes. They are the simplest type of artificial neural network, and there are no loops in the network; data is input, processed through the layers of neurons, and results in an output. The absence of a feedback loop allows for simpler implementations and makes them suitable for tasks that require generalization from the data.

In mathematical terms, each neuron in a feed-forward neural network calculates a weighted sum of its inputs, adds a bias term and then applies an activation function. Given a neuron $j$ in layer $l$, the output of the neuron, $o_{j}^{l}$, is given by:

Weighted sum of inputs: $z_{j}^{l} = \sum_{i} w_{i,j}^{l} \cdot o_{i}^{l-1} + b_{j}^{l}$
Apply activation function: $o_{j}^{l} = f(z_{j}^{l})$
Here, $w_{i,j}^{l}$ is the weight of the connection from neuron $i$ in layer $l-1$ to neuron $j$ in layer $l$, $o_{i}^{l-1}$ is the output of neuron $i$ in layer $l-1$, $b_{j}^{l}$ is the bias of neuron $j$ in layer $l$, and $f$ is the activation function. This process is repeated for every layer in the network until the output layer is reached. The output layer can apply a specialized function, such as softmax for classification tasks, to produce final probabilities for each class.

#### *Convolutional Neural Networks*:
Convolutional Neural Networks (CNNs) are a class of deep learning models most commonly applied to analyzing visual data, although they can also be used on other types of data like NLP. CNNs are designed automatically and adaptively to learn spatial hierarchies of features from the input data. They are characterized by their use of convolutional layers, which apply a series of filters to the input data to extract and learn higher-level features.

In mathematical terms, the key operation in the CNN is the convolution operation. Given an input feature map $I$, a filter $w$ of size $k \times k$ (in 2D), the output feature map $O$ is computed as:

$O_{xy} = \sum_{i=1}^{k} \sum_{j=1}^{k} I_{x+i,y+j} \cdot w_{ij}$

This operation is performed for every position $x, y$ on the input feature map and for every filter in the convolutional layer. This results in a set of output feature maps, one for each filter. Each filter learns to recognize a different feature in the input.

After the convolutional layers, the feature maps are usually passed through a non-linear activation function, often the Rectified Linear Unit (ReLU). Following this, pooling layers (e.g., max pooling or average pooling) are often used to down-sample the feature maps and reduce the computational complexity for the network. Finally, fully connected layers are often used at the end of the network to perform high-level reasoning and make the final classification decision. The output of the network is typically a softmax function that gives the probability distribution over the classes.

#### Precision-Recall Curves:

Precision-Recall curves are used in binary classification to evaluate a model's performance under different probability thresholds. Precision (also known as the positive predictive value) is the proportion of true positive instances among all instances that the model predicted as positive. Recall (also known as sensitivity or true positive rate) is the proportion of true positive instances that were correctly identified by the model. The Precision-Recall curve is particularly useful when dealing with imbalanced datasets, where the number of instances of one class significantly outweighs the other. It provides a more informative picture of an algorithm’s performance by showing the trade-off between precision and recall for different threshold settings. This allows us to choose a threshold that balances precision and recall in a way that makes sense for the particular problem we are solving.

#### Receiver Operating Characteristic (ROC) curves:

An ROC curve is a graphical representation that illustrates the diagnostic ability of a binary classifier system as its discrimination threshold is varied. The curve is created by plotting the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings. The line of no-discrimination represents a random classifier, that is, a classifier that makes decisions randomly with a probability of 50%, without any knowledge or using any features of the data. It is essentially like flipping a coin. If the classifier is no better than random guessing, its ROC curve will lie along the line of no-discrimination. A good classifier will have its ROC curve towards the top left corner, indicating a higher true positive rate and lower false positive rate. Therefore, the area under the ROC curve (also known as AUC-ROC) provides a measure of the model's ability to distinguish between positive and negative classes. The closer the curve is to the top left corner (or the larger the area under the curve), the better the classifier.

#### Confusion Matrix:

A confusion matrix is a specific table layout that visualizes the performance of a supervised learning algorithm. Each row of the matrix represents the instances in a predicted class while each column represents the instances in an actual class. It is a powerful tool for descriptive model evaluation as it shows the true positives, true negatives, false positives, and false negatives, which are crucial for understanding the performance of a classifier. Beyond raw model accuracy, a confusion matrix provides insight into the types of errors the model is making, which can inform further refinements to the model or the choice of model to use. The matrix is particularly useful for binary classification problems.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import re
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
import string
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, SpatialDropout1D, Dense, Embedding, Flatten, Conv1D, MaxPooling1D
from keras.optimizers import RMSprop
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve, average_precision_score
import pickle
import keras
from gensim.models import Word2Vec
from keras.initializers import Constant 
from sklearn.model_selection import train_test_split
```

```python
# Setting global variables
stemmer = SnowballStemmer("english")
stopword=set(stopwords.words('english')) 
```

```python
def load_and_process_data():
    """
    Load and process the Twitter and offensive datasets.
    """
    # Loading data
    df_twitter = pd.read_csv('train.csv')
    df_offensive = pd.read_csv("labeled_data.csv")

    # Basic preprocessing
    df_twitter.drop('id',axis=1,inplace=True) # removing unnecessary columns
    df_offensive = df_offensive[['tweet', 'class']] # keeping only relevant columns
    df_offensive.rename(columns ={'class':'label'}, inplace = True) # renaming for consistency
    df_offensive["label"].replace({0: 1, 2: 0}, inplace=True) # relabeling

    # Combine both dataframes
    df = pd.concat([df_twitter, df_offensive]) 
    df['tweet'] = df['tweet'].apply(clean_text) # applying text cleaning
    
    return df


def clean_text(text):
    """
    Preprocess a given text. Returns cleaned and preprocessed text.
    """
    text = str(text).lower() # convert to lowercase
    text = re.sub('\[.*?\]|https?://\S+|www\.\S+|<.*?>+', '', text) # remove URLs and HTML tags
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # remove punctuation
    text = re.sub('\n|\w*\d\w*', '', text) # remove newline characters and words with numbers
    text = [stemmer.stem(word) for word in text.split(' ') if word not in stopword] # apply stemming and stopword removal
    text = " ".join(text) # join words back into a string
    return text


def create_word2vec_model(texts, vector_dim=100, window_size=5, min_count=1, epochs=100):
    """
    Create a Word2Vec model.
    """
    # Preparing corpus for Word2Vec
    corpus = [nltk.word_tokenize(text) for text in texts]

    # Training Word2Vec model
    w2v_model = Word2Vec(sentences=corpus, vector_size=vector_dim, window=window_size, min_count=min_count, workers=4)
    w2v_model.train(corpus, total_examples=len(corpus), epochs=epochs)

    return w2v_model


def define_lstm_model_w2v(w2v_model, tokenizer, max_words, max_len):
    """
    Define a LSTM model with Word2Vec embeddings.
    """
    # Preparing embedding matrix
    embedding_matrix = np.zeros((max_words, w2v_model.vector_size))
    for word, i in tokenizer.word_index.items():
        if word in w2v_model.wv and i < max_words:
            embedding_matrix[i] = w2v_model.wv[word] # get the vector of the word from the Word2Vec model

    # Defining model
    model = Sequential()
    model.add(Embedding(max_words, w2v_model.vector_size, embeddings_initializer=Constant(embedding_matrix), 
                        input_length=max_len, trainable=False)) # add embedding layer with Word2Vec weights
    model.add(SpatialDropout1D(0.2)) # add dropout for regularization
    model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2)) # add LSTM layer
    model.add(Dense(1, activation='sigmoid')) # output layer
    model.compile(loss='binary_crossentropy', optimizer=RMSprop(), metrics=['accuracy']) # compile model
    
    return model


def define_ffnn_model_w2v(w2v_model, tokenizer, max_words, max_len):
    """
    Define a Feed-Forward Neural Network (FFNN) model with Word2Vec embeddings.
    """
    # Preparing embedding matrix
    embedding_matrix = np.zeros((max_words, w2v_model.vector_size))
    for word, i in tokenizer.word_index.items():
        if word in w2v_model.wv and i < max_words:
            embedding_matrix[i] = w2v_model.wv[word]

    # Defining model
    model = Sequential()
    model.add(Embedding(max_words, w2v_model.vector_size, embeddings_initializer=Constant(embedding_matrix), 
                        input_length=max_len, trainable=False)) 
    model.add(Flatten())
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=RMSprop(), metrics=['accuracy'])
    
    return model


def define_cnn_model_w2v(w2v_model, tokenizer, max_words, max_len):
    """
    Define a Convolutional Neural Network (CNN) model with Word2Vec embeddings.
    """
    # Preparing embedding matrix
    embedding_matrix = np.zeros((max_words, w2v_model.vector_size))
    for word, i in tokenizer.word_index.items():
        if word in w2v_model.wv and i < max_words:
            embedding_matrix[i] = w2v_model.wv[word]

    # Defining model
    model = Sequential()
    model.add(Embedding(max_words, w2v_model.vector_size, embeddings_initializer=Constant(embedding_matrix), 
                        input_length=max_len, trainable=False)) 
    model.add(Conv1D(128, 5, activation='relu'))
    model.add(MaxPooling1D(pool_size=4))
    model.add(LSTM(100))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=RMSprop(), metrics=['accuracy'])
    
    return model

def train_and_evaluate_model(model, x_train, y_train):
    """
    Compiles, trains, and evaluates a given model using training data.

    """
    model.compile(loss='binary_crossentropy', optimizer=RMSprop(), metrics=['accuracy'])
    
    callbacks = [EarlyStopping(monitor='val_loss', patience=2),
                 ModelCheckpoint(filepath='best_model.h5', monitor='val_loss', save_best_only=True)]
                 
    history = model.fit(x_train, y_train, batch_size=128, epochs=10, callbacks=callbacks,
                        validation_split=0.2, shuffle=True, verbose=2)
    return history


def plot_precision_recall_curve(y_test, y_score, model_name):
    """
    Visualize precision-recall curves.
    """
    precision, recall, _ = precision_recall_curve(y_test, y_score)
    average_precision = average_precision_score(y_test, y_score)
    
    plt.step(recall, precision, where='post')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title(f'{model_name}: Precision-Recall curve: AP={average_precision:0.2f}')
    plt.show()

def plot_roc_curve(y_test, y_score, model_name):
    """
    Visualize Receiver Operating Characteristic (ROC) curves; the dotted line represents the "line of 
    no-discrimination." This line is a diagonal line from the bottom left to the top right of the plot.
    """
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name}: Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

The hyperparameters in the function define_lstm_model_w2v were chosen based on a few factors, including the size of the dataset, the complexity of the task, and the computational resources available. The max_words parameter determines the size of the vocabulary that the model will use. This parameter was chosen to be 50,000, which is a relatively large number that should be sufficient to capture the most common words in the dataset. The max_len parameter determines the maximum length of a tweet that the model will accept. This parameter was chosen to be 300, which is a relatively long tweet. This should be sufficient to capture the most important information in a tweet, while also preventing the model from being overwhelmed by very long tweets. The 100 parameter in the LSTM(100) layer determines the number of units in the LSTM layer. This parameter was chosen based on a trial-and-error process, and it was found that 100 units provided a good balance between accuracy and computational resources. The 0.2 parameter in the SpatialDropout1D(0.2) layer determines the dropout rate. Dropout is a regularization technique that helps to prevent overfitting. The dropout rate of 0.2 was chosen based on a trial-and-error process, and it was found that this value helped to improve the accuracy of the model without significantly impacting the computational resources required.

The hyperparameters in the function define_ffnn_model_w2v were chosen based on the same factors as the hyperparameters in the function define_lstm_model_w2v. The max_words and max_len parameters were chosen to be the same as in the define_lstm_model_w2v function. The 50 parameter in the Dense(50) layer determines the number of units in the first hidden layer. This parameter was chosen based on a trial-and-error process, and it was found that 50 units provided a good balance between accuracy and computational resources.

The hyperparameters in the function define_cnn_model_w2v were chosen based on the same factors as the hyperparameters in the function define_lstm_model_w2v, with the addition of the 128 parameter in the Conv1D(128, 5, activation='relu') layer. This parameter determines the number of filters that will be used in the convolutional layer. The value of 128 was chosen based on a trial-and-error process, and it was found that this value provided a good balance between accuracy and computational resources.

## Results
This section will present the results of the machine learning model performances. It will highlight their accuracy scores, confusion matrices, and classification reports, providing a comprehensive comparison across the different models.



```python
# Main function
def main():
    df = load_and_process_data()

    # Visualizing the data
    sns.countplot(df['label'])
    plt.show()

    x=df['tweet']
    y=df['label']
    
    x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
    
    max_words = 50000
    max_len = 300
    tokenizer = Tokenizer(num_words=max_words)
    tokenizer.fit_on_texts(x_train)
    sequences = tokenizer.texts_to_sequences(x_train)
    sequences_matrix = pad_sequences(sequences,maxlen=max_len)

    # Create Word2Vec model
    w2v_model = create_word2vec_model(x_train)

    lstm_model_w2v = define_lstm_model_w2v(w2v_model, tokenizer, max_words, max_len)
    ffnn_model_w2v = define_ffnn_model_w2v(w2v_model, tokenizer, max_words, max_len)
    cnn_model_w2v = define_cnn_model_w2v(w2v_model, tokenizer, max_words, max_len)
    
    stop = EarlyStopping(monitor='val_accuracy', mode='max', patience=5)
    checkpoint_lstm = ModelCheckpoint('lstm_model-{epoch:03d}-{accuracy:03f}-{val_accuracy:03f}.h5', verbose=1, monitor='val_accuracy', mode='max', save_best_only=True)
    checkpoint_ffnn = ModelCheckpoint('ffnn_model-{epoch:03d}-{accuracy:03f}-{val_accuracy:03f}.h5', verbose=1, monitor='val_accuracy', mode='max', save_best_only=True)
    checkpoint_cnn = ModelCheckpoint('cnn_model-{epoch:03d}-{accuracy:03f}-{val_accuracy:03f}.h5', verbose=1, monitor='val_accuracy', mode='max', save_best_only=True)

    lstm_model_w2v.fit(sequences_matrix, y_train, batch_size=32, epochs=10, validation_split=0.2, callbacks=[stop, checkpoint_lstm])
    ffnn_model_w2v.fit(sequences_matrix, y_train, batch_size=32, epochs=10, validation_split=0.2, callbacks=[stop, checkpoint_ffnn])
    cnn_model_w2v.fit(sequences_matrix, y_train, batch_size=32, epochs=10, validation_split=0.2, callbacks=[stop, checkpoint_cnn])

    # Save tokenizer and model
    with open('lstm_model_w2v_tokenizer.pickle', 'wb') as handle:
        pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
        
    lstm_model_w2v.save('lstm_model_w2v.h5')
    ffnn_model_w2v.save('ffnn_model_w2v.h5')
    cnn_model_w2v.save('cnn_model_w2v.h5')

    # Test the model
    test_sequences = tokenizer.texts_to_sequences(x_test)
    test_sequences_matrix = pad_sequences(test_sequences,maxlen=max_len)

    y_score_lstm = lstm_model_w2v.predict(test_sequences_matrix)
    y_score_ffnn = ffnn_model_w2v.predict(test_sequences_matrix)
    y_score_cnn = cnn_model_w2v.predict(test_sequences_matrix)

    plot_precision_recall_curve(y_test, y_score_lstm, 'LSTM Model')
    plot_roc_curve(y_test, y_score_lstm, 'LSTM Model')

    plot_precision_recall_curve(y_test, y_score_ffnn, 'FFNN Model')
    plot_roc_curve(y_test, y_score_ffnn, 'FFNN Model')

    plot_precision_recall_curve(y_test, y_score_cnn, 'CNN Model')
    plot_roc_curve(y_test, y_score_cnn, 'CNN Model')

    y_pred_lstm = (y_score_lstm > 0.5).astype("int32")
    y_pred_ffnn = (y_score_ffnn > 0.5).astype("int32")
    y_pred_cnn = (y_score_cnn > 0.5).astype("int32")

    print("Results for LSTM model with Word2Vec embeddings:")
    print(classification_report(y_test, y_pred_lstm))
    print(confusion_matrix(y_test, y_pred_lstm))

    print("\nResults for FFNN model with Word2Vec embeddings:")
    print(classification_report(y_test, y_pred_ffnn))
    print(confusion_matrix(y_test, y_pred_ffnn))

    print("\nResults for CNN model with Word2Vec embeddings:")
    print(classification_report(y_test, y_pred_cnn))
    print(confusion_matrix(y_test, y_pred_cnn))

    plot_confusion_matrix(y_test, y_pred_lstm, title='LSTM Model')
    plot_confusion_matrix(y_test, y_pred_ffnn, title='FFNN Model')
    plot_confusion_matrix(y_test, y_pred_cnn, title='CNN Model')

def plot_confusion_matrix(y_test, y_pred, title):
    """
    Visualize confusion matrix.
    """
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {title}')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

if __name__ == "__main__":
    main()

```

## Discussion

The goal of this project was to perform text classification on a given dataset, using three different neural network architectures – Long Short Term Memory (LSTM), Feed Forward Neural Network (FFNN) and Convolutional Neural Network (CNN) – with Word2Vec embeddings. The dataset used for the project was a collection of text documents classified into two classes: 0 and 1. The task was to develop models that could correctly classify the text into these two categories. To achieve this, a supervised learning approach was used. The data was split into training and validation sets, with the models trained on the former and evaluated on the latter.

The processing routines utilized in this project primarily concentrated on data preprocessing, feature extraction, model definition, and model training and evaluation. The primary preprocessing routine was load_and_process_data, which was extended to handle more complex preprocessing tasks. It loaded the tweet datasets, dropped irrelevant columns, and merged the datasets. The function also took care of removing retweets and duplicate tweets. This was then followed by the function clean_text that was applied to each tweet. This function not only transformed the text to lowercase, removed URLs, punctuation, and digits but also conducted lemmatization and stopword removal, yielding a better data cleaning result. The cleaned data was then transformed into word embeddings using Word2Vec instead of simple tokenization and sequence conversion, providing a richer feature set for model input.

Three functions, define_lstm_model_w2v, define_ffnn_model_w2v, and define_cnn_model_w2v, were utilized to detail the structure of the LSTM, FFNN, and CNN models, respectively. Each model was compiled using binary cross-entropy as the loss function as the optimizer for improved performance. In the main function, each model was trained using the preprocessed data. The best models were saved during training using ModelCheckpoint, and training was ceased when validation accuracy ceased to enhance for a number of epochs using EarlyStopping. Each model's performance was evaluated using a classification report, a confusion matrix, and the new ROC-AUC score, offering precision, recall, F1-score, accuracy, and AUC metrics.

The communication between these processing routines was coordinated within the main function. The preprocessing routine was used to transform raw data into cleaned and feature extracted data, which was passed onto the model definition routines. These routines returned the defined models, which were subsequently trained and evaluated using the preprocessed data. For each model, validation and quality control were guaranteed by evaluation of the performance metrics provided a comprehensive view of how effectively the models could classify offensive tweets. By comparing these metrics, the superior model was identified.

In terms of model performance, the LSTM model achieved the highest validation accuracy of 94.8%, followed by the CNN model with 94.0% and the FFNN model with 91.9%. Interestingly, the LSTM and CNN models outperformed the FFNN model despite being traditionally more computationally intensive and complex. This can be attributed to the ability of LSTM and CNN models to process sequences, making them more suitable for text data. The LSTM's ability to remember long-term dependencies in the text and the CNN's ability to extract local and global features from the text may have contributed to their superior performance.

The LSTM model also demonstrated superior performance in terms of the precision, recall, and f1-score metrics. These results indicate that the LSTM model not only made accurate predictions but also maintained a good balance between precision and recall, ensuring fewer false positives and false negatives. Although the models' performance on the validation set is encouraging, it is important to note that these results are specific to the particular validation set used. The performance of the models in a real-world scenario may vary depending on the characteristics of the unseen data. Another point to note is the overfitting observed in the models, especially the CNN model. Overfitting refers to a situation where a model learns the training data too well, to the point of fitting the noise in the data, and hence performs poorly on unseen data. This was evident in the CNN model where the validation loss began to increase after the third epoch while the training loss continued to decrease. This gap between the training and validation performance suggests that the model may not generalize well to unseen data. In the end, the LSTM model with Word2Vec embeddings achieved the best performance.


## Conclusion

From the analysis of the training logs for the LSTM model, Feed-Forward Neural Network (FFNN), and Convolutional Neural Network (CNN), each model was evaluated based on the same task. The LSTM model exhibited superior performance with a peak validation accuracy of 94.84% in the 6th epoch, demonstrating its proficiency in managing sequential data, a characteristic that is crucial in Natural Language Processing (NLP) tasks. On the other hand, the FFNN model achieved a maximum validation accuracy of 91.98% in the 2nd epoch, while the CNN model performed slightly better, reaching a peak validation accuracy of 94.05% in the 3rd epoch. However, both the FFNN and CNN models showed a decrease in validation accuracy after their respective peak epochs, suggesting overfitting.

All three models exhibited signs of overfitting, indicated by a lack of improvement or a decrease in validation accuracy after a certain point, along with an increase in validation loss. Overfitting occurs when the model learns too much noise or irrelevant details from the training data, thereby impacting its performance on new, unseen data. To combat overfitting, techniques like early stopping, dropout, or regularization should be implemented. For instance, training could be halted at the 6th epoch for the LSTM model, at the 2nd epoch for the FFNN model, and at the 3rd epoch for the CNN model to prevent the models from overfitting the training data. While the LSTM model outperformed the FFNN and CNN models for this specific task, a critical consideration for future work is to apply techniques to minimize overfitting to enhance the models' ability to generalize to new data.

In the future, several approaches could be explored to enhance the performance of the models used in this project. One way could be tuning the hyperparameters of the models more efficiently, perhaps using techniques like grid search or random search to find an optimal combination. This could potentially improve the learning capabilities of the models. Additionally, it may be beneficial to explore more sophisticated regularization techniques to combat overfitting. For example, introducing variations of dropout or implementing weight decay regularization could help the models generalize better. Experimenting with different network architectures could also be an interesting direction. For instance, hybrid models combining aspects of CNNs and LSTMs might leverage the strengths of both models.

Regarding word embeddings, while Word2Vec has been used in this project, other types of embeddings such as GloVe could also be explored (I had done this and it indeed yielded better results, but including it here goes beyond the scope of this project). GloVe (Global Vectors for Word Representation) is a word embedding method that leverages both local and global statistical information by aggregating global word-word co-occurrence statistics from a corpus. Unlike Word2Vec, which only considers local context and can sometimes miss out on wider semantic relationships, GloVe captures both local context and global statistics, which might give it an edge in capturing semantics more comprehensively. Additionally, considering even more advanced, context-aware embeddings like BERT might provide superior results, as these methods not only consider global co-occurrence statistics but also account for the context in which words appear. However, these methods are computationally more intensive and may require additional resources to implement.

In the end, this project offers valuable insights into the application of different neural network architectures for text classification tasks. These findings contribute to a deeper understanding of the strengths and limitations of each model type and provide guidance for future studies in the field of natural language processing. Despite the challenges identified, the work done here provides a thoroughgoing basis for further exploration and advancement in the field.
